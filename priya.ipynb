```python
x = []
y = []
with open('Churn_Modelling.csv', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
    for row in plots:
        x.append((row[0]))
        y.append((row[1]))

plt.plot(x,y, label = 'Loaded file...!!!')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph')
plt.legend()
plt.show()
```
